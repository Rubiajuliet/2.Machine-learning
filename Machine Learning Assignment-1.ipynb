```python
#pandas is a globalised class having several functions used to handle tables-
#as command is used to change a big command of programme into simpler form
```

```python
import pandas as pd
```

```python
#data bcollection and preprocessing of data
#the dataset file should be stored in the respective folder using.csv(comma separated value)
#have to use the exact spelling of the file for this command
#calling a function (read_csv) from pandas(pd) with the parameter of our dataset file name within double quotation
```

```python
dataset=pd.read_csv("Salary_Data.csv")
dataset
```

```python
#splitting input and output parameters
#assigning a variable name to input i.e independent variable using the exact colomn name which are present in the dataset
#the column name should given in double quotation
#used to assign a column as a variable
```

```python
independent=dataset[["YearsExperience"]]
independent
```

```python
#assigning a variable name to output i.e dependent variable which are present in the dataset
#the column name should given in double quotation
```

```python
dependent=dataset[["Salary"]]
dependent
```

```python
#splitting train and test set
#sklearn is a globalised library contains more functions used for Machine learning
#train_test_split is a funtion name under model_selection(classname) in sklearn used to split the data into train and test sets
#here the library is loaded with 4 kinds of variable positions i.e train input,train output,test input,test output
#we have to assign a variable name according to the variable positions loaded in the library
#the parameters(independent,dependent) should mention in the function name-parameterised function
#train set size should be higher than the test set(i.e 70:30 or 80:20)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation
#importing a function named linearregression for SLR model
#assigning a variable name to the function used for further use of calling 
#the first two lines are just a procedure
#.fit command is used for the application of formula (constant value)to the data given in the open bracket()
#calculating the constant value for train input and its respective train output
#once .fit is executed (i.e the constant values are calculated),i.e our model will be created 
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#if you want to know the constant values (slope/weight) we should use the following commands
```

```python
weight=regressor.coef_
weight
```

```python
#if you want to know the constant values (bias/starting point) we should use the following commands
```

```python
bias=regressor.intercept_
bias
```

```python
#applying the test input data into the regressor programme calculation
#to cross check the created model
#create and assign a variable name(predicted output-y_pred) to it so that we can move to evaluation
#.predict is a function used to apply for the cross checling of the test data
#here all the test data will be cross checked with the created learning model
```

```python
y_pred=regressor.predict(x_test)
```

```python
#evaluation Metrics
#calcalation of R2 value by importing r2_score from sklearn library
#r2 value lies between 0-1
#if the r2 value is near to 1 we can conclude our model has better predition
#if the r2 value is near to 0 then its a poor model
#calculating the r2 value for the actual output(y_test) and the predicted output(y_pred)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#saving the created model
#pickle is a globalised function used to save our created model 
```

```python
import pickle
```

```python
#creating a file name to our finalised model
#it should be saved as .sav format and given as a variable
```

```python
filename="finalized_model_linear.sav"
```

```python
#dumping our created model into the newly created .sav file
#wb means write binary
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
